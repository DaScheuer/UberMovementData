```python
import os
import glob
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read every .csv file in the folder with the formatted datasets for each city
cities_dictionary = {}
formatted_cities_list = glob.glob(os.path.join(r'C:\Users\I539797\Documents\Thesis\Datasets\Formatted Datasets', "*.csv"))

for i in range(len(formatted_cities_list)): # Create iterator
    df = pd.read_csv(formatted_cities_list[i]) # Read one .csv file
    cities_dictionary[f'df{i}'] = df # Name the df object 'df + whatever the current count on the iterator is'
    print(formatted_cities_list[i]) # Print each .csv file name to make sure we iterated through every file in the folder
    
    
# Make sure our df is a Pandas DataFrame object        
print(type(df))
```

```python
df0 = cities_dictionary['df0']
df1 = cities_dictionary['df1']
df2 = cities_dictionary['df2']
df3 = cities_dictionary['df3']
df4 = cities_dictionary['df4']
df5 = cities_dictionary['df5']
df6 = cities_dictionary['df6']
df7 = cities_dictionary['df7']
```

```python
df0_avg_travel_times = df0.groupby('Date')['MeanTravelTimeSeconds'].mean()
df1_avg_travel_times = df1.groupby('Date')['MeanTravelTimeSeconds'].mean()
df2_avg_travel_times = df2.groupby('Date')['MeanTravelTimeSeconds'].mean()
df3_avg_travel_times = df3.groupby('Date')['MeanTravelTimeSeconds'].mean()
df4_avg_travel_times = df4.groupby('Date')['MeanTravelTimeSeconds'].mean()
df5_avg_travel_times = df5.groupby('Date')['MeanTravelTimeSeconds'].mean()
df6_avg_travel_times = df6.groupby('Date')['MeanTravelTimeSeconds'].mean()
df7_avg_travel_times = df7.groupby('Date')['MeanTravelTimeSeconds'].mean()


df0_log_travel_times = np.log(df0_avg_travel_times)
df1_log_travel_times = np.log(df1_avg_travel_times)
df2_log_travel_times = np.log(df2_avg_travel_times)
df3_log_travel_times = np.log(df3_avg_travel_times)
df4_log_travel_times = np.log(df4_avg_travel_times)
df5_log_travel_times = np.log(df5_avg_travel_times)
df6_log_travel_times = np.log(df6_avg_travel_times)
df7_log_travel_times = np.log(df7_avg_travel_times)

fig, ax = plt.subplots(figsize=(50,20))

plt.plot(df0['Date'].unique(), df0_log_travel_times)
plt.plot(df1['Date'].unique(), df1_log_travel_times)
plt.plot(df2['Date'].unique(), df2_log_travel_times)
plt.plot(df3['Date'].unique(), df3_log_travel_times)
plt.plot(df4['Date'].unique(), df4_log_travel_times)
plt.plot(df5['Date'].unique(), df5_log_travel_times)
plt.plot(df6['Date'].unique(), df6_log_travel_times)
plt.plot(df7['Date'].unique(), df7_log_travel_times)

plt.title("Log Average Travel Times", fontsize=50)
plt.xticks(fontsize=30, rotation=45)
plt.yticks(fontsize=30)
plt.xlabel("", fontsize=40)
plt.ylabel("Log Average Travel Times in Seconds", fontsize=40)
ax.xaxis.set_major_locator(plt.MaxNLocator(6))
ax.yaxis.set_major_locator(plt.MaxNLocator(7))
plt.show()
```
