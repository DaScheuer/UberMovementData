```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Read san francisco csv
dfSF = pd.read_csv(r'C:\Users\I539797\Documents\GitHub\UberMovementDataDownloadBot\Datasets\Formatted Datasets\formatted_sanfrancisco.csv')

# Read los angeles csv
dfLA = pd.read_csv(r'C:\Users\I539797\Documents\GitHub\UberMovementDataDownloadBot\Datasets\Formatted Datasets\formatted_losangeles.csv')
```

```python
SF_avg_travel_times, LA_avg_travel_times = dfSF.groupby('Date')['MeanTravelTimeSeconds'].mean(), dfLA.groupby('Date')['MeanTravelTimeSeconds'].mean()
SF_log_travel_times, LA_log_travel_times = np.log(SF_avg_travel_times), np.log(LA_avg_travel_times)
```

```python
fig, ax = plt.subplots(figsize=(50,20))

plt.scatter(dfSF['Date'].unique(), SF_log_travel_times)
plt.scatter(dfLA['Date'].unique(), LA_log_travel_times)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(50,20))
plt.scatter(dfSF['Date'].unique(), SF_avg_travel_times)
plt.scatter(dfLA['Date'].unique(), LA_avg_travel_times)
plt.show()
```
